```python
import numpy as np
import matplotlib.pyplot as plt

# constants
alpha0 = 1
sigma0 = 2
L = 1
I0 = 1

# r array
r = np.linspace(0.1*L, L, 500)

# Case (a) r^2
tau_r2 = (alpha0 + sigma0) * r**3 / 3
I_r2 = I0 * np.exp(-tau_r2)

# Case (b) ln(r)
tau_ln = (alpha0 + sigma0) * (r * np.log(r) - r)
I_ln = I0 * np.exp(-tau_ln)

# Plotting
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(r, I_r2, label='Intensity (r)', color='blue')
plt.title('Observed Intensity for $f(r) = r^2$')
plt.xlabel('r',size=15)
plt.ylabel('I(r)',size=15)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(r, I_ln, label='Intensity (ln r)', color='red')
plt.title('Observed Intensity for $f(r) = \ln(r)$')
plt.xlabel('r',size=15)
plt.ylabel('I(r)',size=15)
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
